### hitung integral dan diferensial dari fungsi 'f' untuk interval x

```python
import sympy as sp

def f(x) :
    return np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)

# batas interval
a, b = 0, 10
n = 500
h = (b - a) / n

# titik-titik integral
x_points = np.linspace(a, b, n+1)

# perhitungan luas trapesium manual
areas = []
for i in range(n):
    x0 = x_points[i]
    x1 = x_points[i+1]
    L = (f(x0) + f(x1)) / 2 * h
    areas.append(L)
    print(f"Luas trapesium {i+1} (x={x0:.2f}→{x1:.2f}) = {L:.6f}")

I_num = sum(areas)
print("\nIntegral numerik (Trapesium) =", I_num)

# perhitungan secara teoritis/analitik
x = sp.symbols('x')
f_sym = sp.sin(x) + 0.5*sp.sin(2*x) + (1/sp.Integer(3))*sp.sin(3*x) + (1/sp.Integer(4))*sp.sin(4*x) + (1/sp.Integer(5))*sp.sin(5*x)

I_theo = sp.integrate(f_sym, (x, a, b))
print("Integral teoritis (Analitik)   =", float(I_theo))
```

```python
# interval
x = np.linspace(0, 10, 500)

# batas interval
a, b = 0, 10
n = 500
h = (b - a) / n

# diferensial f'(x)
x_vals = np.linspace(a, b, n+1)
f_vals = f(x_vals)
f_prime_num = np.gradient(f_vals, x_vals)

print("nilai f'(x) numerik di titik pertama:", f_prime_num[0])
print("nilai f'(x) numerik di titik terakhir:", f_prime_num[-1])
```

### hitung secara terpisah bagian integral diatas dan dibawah y=0

```python
# interval
x = np.linspace(0, 10, 500)

# fungsi sinusoidal 
f = (np.sin(x)
     + 0.5 * np.sin(2*x)
     + (1/3) * np.sin(3*x)
     + (1/4) * np.sin(4*x)
     + (1/5) * np.sin(5*x))

# integral numerik total
I_total = np.trapz(f, x)

# memisahkan bagian positif negatif atau atas dan bawah y=0
f_positif = np.where(f > 0, f, 0)
f_negatif = np.where(f < 0, f, 0)

# memisahkan titik-titik sumbu
I_positif = np.trapz(f_positif, x)
I_negatif = np.trapz(f_negatif, x)
I_negatif_abs = -I_negatif # titik dibawah sumbu

print("Integral total        =", I_total)
print("Integral diatas (>)  =", I_positif)
print("Integral dibawah (<) =", I_negatif, "(signed)")
print("Titik dibawah sumbu  =", I_negatif_abs)
print("Check: I_positif + I_negatif  =", I_positif + I_negatif)
```

### Penjelasan :
* untuk menghasilkan area atas dan bawah :
  1. bagian fungsi yang lebih besar dari nol dianggap sebagai area diatas sumbu/sumbu positif.
  2. bagian fungsi yang lebih kecil dari nol dianggap sebagai area dibawah sumbu/sumbu negatif.
* untuk menghitung luas masing masing daerah sumbu :
  1. "I_positif" digunakan sebagai perhitungan total luas area diatas sumbu.
  2. "I_negatif" digunakan sebagai perhitungan total luas area dibawah sumbu, namun bernilai negatif dimana sesuai dengan definisi integral.
  3. "I_negatif_abs" digunakan sebagai perhitungan luas titik dibawah sumbu dengan niali positif, agar dapat mengetahui besar nilainya secara nyata.

### plot semua fungsi : f, f' dan F (atas dan bawah dengan area warna berbeda)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid

# interval dan fungsi (f)
x = np.linspace(0, 10, 500)
f = (np.sin(x) 
     + 0.5*np.sin(2*x) 
     + (1/3)*np.sin(3*x) 
     + (1/4)*np.sin(4*x) 
     + (1/5)*np.sin(5*x))

# turunan numerik f'
f_prime = np.gradient(f, x)

# integral kumulatif F(x)
F = cumulative_trapezoid(f, x, initial=0)

# - KESELURUHAN VISUALISASI -
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# visualisasi 1 f(x)
axs[0].plot(x, f, label="f(x)")
axs[0].fill_between(x, f, 0, where=(f>0), color="green", alpha=0.3, label="Area > 0/positif")
axs[0].fill_between(x, f, 0, where=(f<0), color="red", alpha=0.3, label="Area < 0/negatif")
axs[0].axhline(0, color='k', linewidth=0.7)
axs[0].set_title("Fungsi f(x) dengan area atas & bawah")
axs[0].legend()

# visualisasi 2 f'(x)
axs[1].plot(x, f_prime, color="orange", label="f'(x)")
axs[1].axhline(0, color='k', linewidth=0.7)
axs[1].set_title("Turunan f'(x)")
axs[1].legend()

# visualisasi 3 F(x) (integral kumulatif)
axs[2].plot(x, F, color="purple", label="F(x)")
axs[2].axhline(0, color='k', linewidth=0.7)
axs[2].set_title("Integral kumulatif F(x)")
axs[2].legend()

plt.tight_layout()
plt.show()
```

### Penjelasan :
* visualisasi 1 f(x):
  1. grafik yang menampilkan fungsi asli,
  2. area diatas sumbu/sumbu positif berwarna hijau, sedangkan area dibawah sumbu/sumbu negatif berwarna merah.
  3. grafik ini menggambarkan bagaimana integral positif dan negatif terbentuk.

* visualisasi 2 f'(x):
  1. grafik yang menampilkan turunan fungsi.
  2. menunjukkan kecepatan perubahan f(x), di mana grafik memotong sumbu x saat fungsi mencapai titik puncak/lembah.

* visualisasi 3 F(x):
  1. grafik yang menampilkan integral kumulatif.
  2. jika nilai f(x) banyak bernilai negatif, maka grafik F(x) akan turun, sebaliknya jika nilai f(x) banyak bernilai positif, maka grafik F(x) akan naik.
